```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Basics

```python
# we will use linspace -> linear space
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
# lets use plotting
plt.plot(x, y)
```

```python
# adding labels and legends
plt.title("Plotting")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

## Subplots in matplotlib

```python
# plt.title("Plotting subplots")
# plt.xlabel("X-axis")
# plt.ylabel("Y-axis")
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.plot(x**2,y**2)

plt.subplot(2,2,3)
# add just below the plot to make it working
plt.plot(x,y)

plt.subplot(2,2,4)
plt.plot(x**3,y**2)
```

## Object oriented way

```python
# we will see way to access canvas so that we can move plot at index 1 -> 3 and soon.
fig = plt.figure()
fig # this is a canvas

# adding axes
axis1 = fig.add_axes([0.1,0.1,0.8,0.8]) # left space, bottom, width, height
axis1.plot(x,y)

axis2 = fig.add_axes([0.2,0.5,0.2,0.2]) # left space, bottom, width, height
axis2.plot(x,y)
```

## Types of plots

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
# no y var, only x data and y works with x accordingly
```

```python
from random import sample
data = sample(range(1, 500), 100)
plt.hist(data)
```

```python
plt.boxplot(x)
# same only one axis
```

## Saving plots

```python
fig = plt.figure()
axis = fig.add_axes([0.1,0.1,0.8,0.8]) 
axis.plot(x,y)
plt.savefig("basicplot")
```
