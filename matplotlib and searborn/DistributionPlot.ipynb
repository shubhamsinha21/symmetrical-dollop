Distribution plots are created for quantative data
lets discuss some of imp ones -

1. histplot/distplot
2. jointplot
3. pairplot
4. rugplot

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df['size'].unique()
```

### HistPlot

```python
sns.histplot(df['total_bill'], bins =10)
# historgram is best for distribution showcase
# bins - width/size of bar
# kde(kernel density estimator) - it says data me skewness kitna hai.
# data left ya right skewed hai
```

```python
sns.histplot(df['total_bill'], bins =10, kde = True)
```

```python
# subplot of both tips and total_bill
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins =10, kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins =10, kde = True)
```

### Joint Plot
- is used to compare the two data 

```python
sns.jointplot(x='total_bill', y ='tip', data = df, kind= "scatter")
```

```python
sns.jointplot(x='total_bill', y ='tip', data = df, kind= "kde")
```

```python
sns.jointplot(x='total_bill', y ='tip', data = df, kind= "kde")
```
