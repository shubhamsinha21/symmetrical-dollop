```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## matrix plot

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Correlation
- bet -1 and +1
- if corr = 0 , not useful for prediction
- A value increases , C also increases -> +ve correlation
- A value decreases, C value increses ->  -ve correlation
- we use heatmap plot to show correlation
- for both categorical and quantitative

## Heatmap

```python
# take numerica data in tips
tipscorr = tips[['total_bill', 'tip', 'size']]
tipscorr
```

```python
tipscorr.corr()
# correlation bet each others
```

```python
sns.heatmap(tipscorr.corr())
```

```python
sns.heatmap(tipscorr.corr(), annot=True)
# shows values of correlation also
```

## Pivot table heatmaps

```python
pvtflight = flights.pivot_table(values ='passengers', index='month', columns='year')
pvtflight
```

```python
sns.heatmap(pvtflight)
# we can add directly data since all data are numerical
```
