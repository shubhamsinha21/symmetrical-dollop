## Categorical Plots

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df.head()
```

### Count plots

```python
sns.countplot(x = df['sex'])
```

```python
# to check how many males and females
sns.countplot(x = df['sex'], hue = df['smoker'])
```

### Bar plots

```python
sns.barplot(x = df['sex'], y = df['total_bill'])

# x-axis has categorical data
# y-axis has quantitative data
```

```python
sns.barplot(x = df['sex'], y = df['total_bill'], estimator = np.sum)
# default estimator is mean
```

```python
sns.barplot(x = df['sex'], y = df['tip'], estimator = np.sum)
# default estimator is mean
```

### Box plots

```python
sns.boxplot(x='tip', y='day', data = df)
```
