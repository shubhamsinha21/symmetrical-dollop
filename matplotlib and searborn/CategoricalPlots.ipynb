## Categorical Plots

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df.head()
```

### Count plots

```python
sns.countplot(x = df['sex'])
```

```python
# to check how many males and females
sns.countplot(x = df['sex'], hue = df['smoker'])
```

### Bar plots

```python
sns.barplot(x = df['sex'], y = df['total_bill'])

# x-axis has categorical data
# y-axis has quantitative data
```

```python
sns.barplot(x = df['sex'], y = df['total_bill'], estimator = np.sum)
# default estimator is mean
```

```python
sns.barplot(x = df['sex'], y = df['tip'], estimator = np.sum)
# default estimator is mean
```

### Box plots

```python
sns.boxplot(x='tip', y='day', data = df, palette='rainbow', hue = df['smoker'])
```

### Violin plots

```python
sns.violinplot(x='day', y='tip', data = df)

# width more has more data
# width less has lesser data
```

### Strip plot and Swarm plot

stripplot
- one variable is categorical
- complement for violin plot

swarmplot
- are similar to stripplots but points are adjusted only along categorical axis
- results as they dont overlap

```python
sns.stripplot(x='day', y='tip', data=df)
```

```python
sns.swarmplot(x='day', y='tip', data=df)
# wider shape where it has more data
```

```python
# combining violin and swarmplot
sns.violinplot(x='tip', y='day', data = df)
sns.swarmplot(x='tip', y='day', data=df)
```
