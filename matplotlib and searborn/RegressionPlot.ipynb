## lmplot()
- regression plot is used in ml so we will not fully discuss what, how and where it i used.
- we will focus on how we can create it and what values can be put inside

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data= tips)

# x -> input helps in predict
# y -> output we have to predict
```

```python
sns.lmplot(x='total_bill', y='tip', data= tips, hue="sex")
```

```python
sns.lmplot(x='total_bill', y='tip', data= tips, hue="sex", markers=['o', 'v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data= tips, 
           hue="sex", markers=['o', 'v'], 
           scatter_kws={'s': 60}
          )
```
