```python
import numpy as np
import pandas as pd
```

Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
# to find null values
df.isna()
```

```python
df.isna().sum()
```

```python
# to find all columns has atleast one non-value ?
df.isna().any()
```

Removing Missing Data

```python
df
```

```python
# removing the data works based on rows
df.dropna() 
# removes the row which has any NaN value
```

```python
df
```

```python
df.dropna(thresh = 3) # it means rows mei atleast 3 non-null values (thresh)
```

Filling the missing Data

```python
df
```

```python
# filling non-null values with zero
df.fillna(0)
```

```python
df
```

```python
# replace null values with specific values as given below
values = {'A':0, 'B':100, 'C': 300, 'D': 400}
df.fillna(value=values)
```

```python
# now, i want to fill the mean values for  NaN vales
df.fillna(df.mean())
```
