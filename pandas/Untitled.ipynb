### Creating series

```python
import numpy as np
import pandas as pd
```

labels are called index here..

```python
labels = ['a', 'b', 'c'] # labels
my_list = [10,20,30] # list
arr = np.array([10,20,30]) # array
d = {1:10, 2:20} # dict
```

```python
pd.Series(my_list)
# labels/index | values/datapoints
```

```python
# setting custom labels
pd.Series(my_list, index=labels)
```

```python
# can we create Series with an array
pd.Series(arr)
```

- series are 1d so you must keep array as vectors (1d)
- in case of dict, labels are defined itself using key
