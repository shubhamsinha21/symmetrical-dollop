### Pandas

- is a 1-d labelled array capable of holding any data type
- the axis labels are collectively called the index

### Creating series

```python
import numpy as np
import pandas as pd
```

- labels are called index here..

```python
labels = ['a', 'b', 'c'] # labels
my_list = [10,20,30] # list
arr = np.array([10,20,30]) # array
d = {1:10, 2:20} # dict
```

```python
pd.Series(my_list)
# labels/index | values/datapoints
```

```python
# setting custom labels
pd.Series(my_list, index=labels)
```

```python
# can we create Series with an array
pd.Series(arr)
```

- series are 1d so you must keep array as vectors (1d)
- in case of dict, labels are defined itself using key

### DataFrames

1. Creating a dataframe

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df
```

```python
# dataframe of a list
# 2d array
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
df2

columns = ["Name", "Age", "City", "Salary"]
df2 = pd.DataFrame(data_list, columns=columns)
df2
```

2. Selection and Indexing of columns

```python
df2['Name']
```

```python
# targetting 2 columns at once
df2[['Name', 'City']]
```

3. Creating a new columns

```python
df2['Designation'] = ["Doctor", "Engineer", "Tailor", "Clerk"]
df2
```
