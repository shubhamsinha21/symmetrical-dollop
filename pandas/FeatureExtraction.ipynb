Feature extraction

```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ')':
            return data 
        if check == True:
            data += i     
        if i == '(':
            check = True  
```

```python
df["Episodes"] = df['Title'].apply(extract_episodes)
df.head()
```

```python
# remove eps
df['Episodes'] = df['Episodes'].str.replace(" eps", "")
df.head()
```

```python
# convert  episodes string type -> integer type
df.loc[0]['Episodes']
```

```python
df['Episodes'] = df['Episodes'].astype(int)
df.head()
```

```python
# make a new column for timestamp
df.loc[3]['Title']
```

```python
def extract_timestamp(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1, i+20):
                data += txt[j]
    return data
```

```python
# make a column for no of months
from datetime import datetime
def extract_months(txt):

    start, end = txt.split(" - ")

    start_date = datetime.strptime(start, "%b %Y")
    end_date = datetime.strptime(end, "%b %Y")

    months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)

    return months
```

```python
df["Timestamps"] = df['Title'].apply(extract_timestamp)
df.head()
```

```python
type(df.loc[0]['Timestamps'])
```

```python
df['Months'] = df['Timestamps'].apply(extract_months).astype(int)
df.head()
```
